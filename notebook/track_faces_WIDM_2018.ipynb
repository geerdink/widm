### This demo shows how to detect faces from video and displays result in realtime

You need OpenCV installed to run this example. To install it, run ___pip install opencv-python___


```python
%pylab inline 
import face_recognition
import cv2
import matplotlib.patches as patches
from IPython.display import clear_output
from matplotlib.pyplot import imshow
import matplotlib.pylab as plt
```

```python
# Load some sample pictures and learn how to recognize them.
en_image = face_recognition.load_image_file("../bella-voorstel.jpg")
en_face_encoding = face_recognition.face_encodings(en_image)[0]

ei_image = face_recognition.load_image_file("../emilio-voorstel.jpg")
ei_face_encoding = face_recognition.face_encodings(ei_image)[0]

ji_image = face_recognition.load_image_file("../jan-voorstel.jpg")
ji_face_encoding = face_recognition.face_encodings(ji_image)[0]

ml_image = face_recognition.load_image_file("../jean-marc-voorstel.jpg")
ml_face_encoding = face_recognition.face_encodings(ml_image)[0]

ns_image = face_recognition.load_image_file("../loes-voorstel.jpg")
ns_face_encoding = face_recognition.face_encodings(ns_image)[0]

ne_image = face_recognition.load_image_file("../olcay-voorstel.jpg")
ne_face_encoding = face_recognition.face_encodings(ne_image)[0]

rl_image = face_recognition.load_image_file("../ron-voorstel.jpg")
rl_face_encoding = face_recognition.face_encodings(rl_image)[0]

rt_image = face_recognition.load_image_file("../ruben-voorstel.jpg")
rt_face_encoding = face_recognition.face_encodings(rt_image)[0]

sh_image = face_recognition.load_image_file("../simone-voorstel.jpg")
sh_face_encoding = face_recognition.face_encodings(sh_image)[0]

sn_image = face_recognition.load_image_file("../stine-voorstel.jpg")
sn_face_encoding = face_recognition.face_encodings(sn_image)[0]

known_faces = [
    en_face_encoding,
    ei_face_encoding,
    ji_face_encoding,
    ml_face_encoding,
    ns_face_encoding,
    ne_face_encoding,
    rl_face_encoding,
    rt_face_encoding,
    sh_face_encoding,
    sn_face_encoding,
]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
frame_number = 0
```

```python
bella = 0
emilio = 0
jan = 0
jean_marc = 0
loes = 0
olcay = 0
ron = 0
ruben = 0
simone = 0
stine = 0
```

```python
# Create an output movie file (make sure resolution/frame rate matches input video!)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_movie = cv2.VideoWriter('output4.avi', fourcc, 25, (640, 360))
```

```python
# Loading video for face detection
# afl1 : bb_AT_2091347.m4v
# afl2 : bb_AT_2091348.m4v
# afl3 : bb_AT_2091349.m4v
# afl4 : bb_AT_2091350.m4v
video_capture = cv2.VideoCapture("../bb_AT_2091350.m4v")
length = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
print(length)
```

```python
frame_count = 0
a = False

while video_capture.isOpened():    
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Bail out when the video file ends
    if not ret:
        video_capture.release()
        break
        
    # We will search face in every 15 frames to speed up process.
    frame_count += 1
    if frame_count % 500 == 0:
        print('{}/{}'.format(frame_count,length))
    if frame_count % 15 == 0:    
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)      

        # Find all the faces and face encodings in the current frame of video
        rgb_frame = frame[:, :, ::-1]
        face_locations = face_recognition.face_locations(rgb_frame)
        face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

        face_names = []
        for face_encoding in face_encodings:
            # See if the face is a match for the known face(s)
            match = face_recognition.compare_faces(known_faces, face_encoding, tolerance=0.50)

            # If you had more than 2 faces, you could make this logic a lot prettier
            # but I kept it simple for the demo
            name = None
            if match[0]:
                bella += 1 
                name = 'bella'
            elif match[1]:
                emilio += 1
                name = 'emilio'                
            elif match[2]:                
                jan += 1
                name = 'jan'
            elif match[3]:                
                jean_marc += 1
                name = 'jean_marc'
            elif match[4]:                
                loes += 1
                name = 'loes'
            elif match[5]:                
                olcay += 1
                name = 'olcay'
            elif match[6]:                
                ron += 1
                name = 'ron'
            elif match[7]:                
                ruben += 1
                name = 'ruben'
            elif match[8]:            
                simone += 1
                name = 'simone'
            elif match[9]:                
                stine += 1 
                name = 'stine'
            
            face_names.append(name)
        
        for (top, right, bottom, left), name in zip(face_locations, face_names):
            if not name:
                continue
                
            cv2.rectangle(frame, (left, top), (right, bottom), (0,0, 255),2)
            
            cv2.rectangle(frame, (left, bottom -25), (right, bottom), (0,0,255), cv2.FILLED)
            font = cv2.FONT_HERSHEY_DUPLEX
            cv2.putText(frame, name, (left + 6, bottom -6), font, 0.5, (255, 255, 255), 1)
            
        #print('writing frame {} / {}'.format(frame_number, length))
        output_movie.write(frame)
cv2.destroyAllWindows()
```
