### This demo shows how to detect faces from video and displays result in realtime

You need OpenCV installed to run this example. To install it, run ___pip install opencv-python___


```python
%pylab inline 
import face_recognition
import cv2
import matplotlib.patches as patches
from IPython.display import clear_output
from matplotlib.pyplot import imshow
import matplotlib.pylab as plt
```

```python
# Load some sample pictures and learn how to recognize them.
image_1 = face_recognition.load_image_file("../img/2022/avatar_arno_nieuw.jpeg")
face_encoding_1 = face_recognition.face_encodings(image_1)[0]

image_2 = face_recognition.load_image_file("../img/2022/avatar_everon_nieuw.jpeg")
face_encoding_2 = face_recognition.face_encodings(image_2)[0]

image_3 = face_recognition.load_image_file("../img/2022/avatar_fresia_nieuw.jpeg")
face_encoding_3 = face_recognition.face_encodings(image_3)[0]

image_4 = face_recognition.load_image_file("../img/2022/avatar_glen_nieuw.jpeg")
face_encoding_4 = face_recognition.face_encodings(image_4)[0]

image_5 = face_recognition.load_image_file("../img/2022/avatar_hila_nieuw.jpeg")
face_encoding_5 = face_recognition.face_encodings(image_5)[0]

image_6 = face_recognition.load_image_file("../img/2022/avatar_kim-lian_nieuw.jpeg")
face_encoding_6 = face_recognition.face_encodings(image_6)[0]

image_7 = face_recognition.load_image_file("../img/2022/avatar_laetitia_nieuw.jpeg")
face_encoding_7 = face_recognition.face_encodings(image_7)[0]

image_8 = face_recognition.load_image_file("../img/2022/avatar_sahil_nieuw.jpeg")
face_encoding_8 = face_recognition.face_encodings(image_8)[0]

image_9 = face_recognition.load_image_file("../img/2022/avatar_suzanne_nieuw.jpeg")
face_encoding_9 = face_recognition.face_encodings(image_9)[0]

image_10 = face_recognition.load_image_file("../img/2022/avatar_thomas_nieuw2.jpg")
face_encoding_10 = face_recognition.face_encodings(image_10)[0]

image_11 = face_recognition.load_image_file("../img/2022/avatar_welmoed_nieuw.jpeg")
face_encoding_11 = face_recognition.face_encodings(image_11)[0]



known_faces = [  
    face_encoding_1,
    face_encoding_2,
    face_encoding_3,
    face_encoding_4,
    face_encoding_5,
    face_encoding_6,
    face_encoding_7,
    face_encoding_8,
    face_encoding_9,
    face_encoding_10,
    face_encoding_11    
]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
frame_number = 0
```

```python
p1 = 0
p2 = 0
p3 = 0
p4 = 0
p5 = 0
p6 = 0
p7 = 0
p8 = 0
p9 = 0
p10 = 0
p11 = 0
```

```python
# Loading video for face detection
# afl1 : S2022E01
# afl2 : S2022E02
# afl3 : S2022E03
eps = 'S22E01'
# Create an output movie file (make sure resolution/frame rate matches input video!)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_movie = cv2.VideoWriter(f'../episodes/{eps}_tracked.avi', fourcc, 25, (1280, 720))

video_capture = cv2.VideoCapture(f"../episodes/{eps}.mp4")
length = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
print(length)

frame_count = 0
a = False

while video_capture.isOpened():    
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Bail out when the video file ends
    if not ret:
        video_capture.release()
        break
        
    # We will search face in every 15 frames to speed up process.
    frame_count += 1
    if frame_count % 500 == 0:
        print('{}/{}'.format(frame_count,length))
    if frame_count % 15 == 0:    
        #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)      

        # Find all the faces and face encodings in the current frame of video
        rgb_frame = frame[:, :, ::-1]
        face_locations = face_recognition.face_locations(rgb_frame)
        face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

        face_names = []
        for face_encoding in face_encodings:
            # See if the face is a match for the known face(s)
            match = face_recognition.compare_faces(known_faces, face_encoding, tolerance=0.50)

            # If you had more than 2 faces, you could make this logic a lot prettier
            # but I kept it simple for the demo
            name = None
            if match[0]:
                p1 += 1
                name = 'arno'
            elif match[1]:
                p2 += 1
                name = 'everon'
            elif match[2]:
                p3 += 1
                name = 'fresia'
            elif match[3]:
                p4 += 1
                name = 'glen'
            elif match[4]:
                p5 += 1
                name = 'hila'
            elif match[5]:
                p6 += 1
                name = 'kim-lian'
            elif match[6]:
                p7 += 1
                name = 'laetitia'
            elif match[7]:
                p8 += 1
                name = 'sahil'
            elif match[8]:
                p9 += 1
                name = 'suzanne'
            elif match[9]:
                p10 += 1
                name = 'thomas'
            elif match[10]:
                p11 += 1
                name = 'welmoed'                
            
            face_names.append(name)
        
        for (top, right, bottom, left), name in zip(face_locations, face_names):
            if not name:
                continue
                
            cv2.rectangle(frame, (left, top), (right, bottom), (0,0, 255),2)
            
            cv2.rectangle(frame, (left, bottom -25), (right, bottom), (0,0,255), cv2.FILLED)
            font = cv2.FONT_HERSHEY_DUPLEX
            cv2.putText(frame, name, (left + 6, bottom -6), font, 0.5, (255, 255, 255), 1)
            
        #print('writing frame {} / {}'.format(frame_number, length))
        output_movie.write(frame)
cv2.destroyAllWindows()
```
